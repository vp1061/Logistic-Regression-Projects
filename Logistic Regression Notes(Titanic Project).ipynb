```python
def add_age(col_new_age):
    Age = col_new_age[0]
    Pclass = col_new_age[1]
    
    if pd.isnull(Age):
        if Pclass == 1 :
            return 37
        elif Pclass == 2:
            return 29
        elif Pclass == 3:
            return 24
    else:
        return Age
```

```python
train['Age'] = train[['Age','Pclass']].apply(add_age,axis=1)
```

**<div style="font-size:30px;color:blue;">ML Logistic Regression</div>**

```python
#features and train predict
X = trainCln.drop('Survived',axis=1)
y = trainCln['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logModel = LogisticRegression()
```

```python
#to train or fit the data(training)
logModel.fit(X_train,y_train)
```

```python
#predictions
predictions = logModel.predict(X_test)
```

```python
#evaluating the mode mae,mse,rmse
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
